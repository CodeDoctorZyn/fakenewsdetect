```python
df = df.dropna() 
# Step 3: Preprocessing
# Drop any rows with missing values
```

```python
# Separate labels and features
X = df['text']  # Text of the article
y = df['subject']  # Label: 'REAL' or 'FAKE'

```

```python
# Separate labels and features
X = df['text']  # Text of the article
y = df['subject']  # Replace with the column that has "FAKE" or "REAL" labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
# Step 5: Text vectorization
# Using TfidfVectorizer to convert text to numerical features
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
# Step 6: Model Training
# Use PassiveAggressiveClassifier (a good baseline for text classification tasks)
model = PassiveAggressiveClassifier(max_iter=50)
model.fit(tfidf_train, y_train)
```

```python
# Step 7: Model Evaluation
# Predict and calculate accuracy
y_pred = model.predict(tfidf_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Confusion Matrix
confusion = confusion_matrix(y_test, y_pred, labels=['FAKE', 'REAL'])
print("Confusion Matrix:")
print(confusion)
```

```python
# Step 8: Testing the model with a custom input
def predict_news(text):
    transformed_text = tfidf_vectorizer.transform([text])
    prediction = model.predict(transformed_text)
    return prediction[0]
```
